```python
import numpy as np
from numpy import linalg as LA
a = [3,1,1,1,3,2,1,1,4]
A=np.reshape(a,(3,3))
B=[3,0,7]
#B=np.reshape(B,(3,1))
IG=[0.0,0.0,0.0] 
tol = 1e-6
itr=50
#np.transpose(B)
```

```python
def GS(a,b, initial_guess ,tol,itr):
    n=len(a)
    x=initial_guess.copy() 
    #x=initial_guess
    print(a)
    print(b)
    for k in range(1,itr):
        xo=x.copy()
        print(f"Iteration {k}: {xo}")
        for i in range(n):
            sum1=np.dot(a[i,:i],x[:i])
            sum2=np.dot(a[i,(i+1):],xo[(i+1):])
            x[i]=(b[i]-sum1-sum2)/a[i,i]
        if(np.allclose(x,xo,tol)):
            print(f"Solution: {x}")
            print(k)
            break
    return x   
    
    
    
```

```python
y=GS(A,B, IG ,tol,itr)
```
